```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense,Dropout
from sklearn.preprocessing import LabelEncoder
import numpy

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf

seed = 3
numpy.random.seed(seed)
tf.random.set_seed(seed)

df = pd.read_csv('../data/sonar.csv',header=None)
print(df.info())
```

```python
df
```

```python
dataset = df.values
dataset
```

```python
X = dataset[:,0:60].astype(float)
Y_obj = dataset[:,60]
```

```python
X
```

```python
Y_obj
```

```python
# 문자열을 숫자로 변환
e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)
type(Y)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=seed)
```

```python
model = Sequential()
model.add(Dense(50,input_dim=60,activation='relu'))
model.add(Dense(30,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='mean_squared_error',optimizer='adam',metrics=['accuracy'])
model.fit(X_train,Y_train,epochs=130,batch_size=5)
model.save('../data/my_model.h5')
```

```python
del model
model = load_model('../data/my_model.h5')
```

```python
print('\n Test Accuracy: %.4f'%(model.evaluate(X_test,Y_test)[1]))
```

#### Stratified K-Fold

```python
from sklearn.model_selection import StratifiedKFold

dataset = df.values
X = dataset[:,0:60].astype(float)
Y_obj = dataset[:,60]

e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)

n_fold = 10
```

```python
skf = StratifiedKFold(n_splits=n_fold,shuffle=True,random_state=seed)
n_iter = 0
accuracy = []
for train_index,test_index in skf.split(X,Y):
    X_train,X_test = X[train_index],X[test_index]
    Y_train,Y_test = Y[train_index],Y[test_index]
    model = Sequential()
    model.add(Dense(24,input_dim=60,activation='relu'))
    model.add(Dense(10,activation='relu'))
    model.add(Dense(1,activation='sigmoid'))
    model.compile(loss='mean_squared_error',optimizer='adam',metrics=['accuracy'])
    model.fit(X_train,Y_train,epochs=100,batch_size=5)
    k_accuracy = '%.4f'%(model.evaluate(X_test,Y_test)[1])
    accuracy.append(float(k_accuracy))
    
print('\n %.f fold accuracy:'%n_fold,accuracy)
print('\n %.f fold accuracy mean:'%n_fold,np.round(np.mean(accuracy),4))
```
